```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
phone = pd.read_csv('phones.csv')
phone.head()
```

```python
sns.set_style('darkgrid') #! this styles the plot
```

```python
phone['price']= phone.original_price #! this sets a certain column as an index for the dataframe.
phone['newprice'] = phone.discounted_price
phone['memory'] = phone.memory
```

```python
ax = plt.axes(projection='3d')
ax.scatter(phone['price'],phone['newprice'],phone['memory']) #! the plot is drawn without specifieng X,Y,Z.
ax.set_xlabel('price') #! sets labels for the plot.
ax.set_ylabel('new Price')
ax.set_zlabel('memory')
```

```python
sns.set_style('whitegrid')
ax = plt.axes(projection='3d')
# ! the plot is drawn without specifieng X,Y,Z.
ax.scatter(phone['price'], phone['newprice'], phone['memory'])
ax.set_xlabel('price')  # ! sets labels for the plot.
ax.set_ylabel('new Price')
ax.set_zlabel('memory')
```
