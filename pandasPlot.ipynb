```python
import numpy as np
import pandas as pd
```

```python
phones = pd.read_csv('phones.csv')
phones.set_index(['brand','model'],inplace = True) #! sets indexes for the dataframe, here I set multiple inexes
phones.index
```

```python
phones.loc['APPLE'] #! now you can search using the row as an index
```

```python
phones.plot(legend=True,title = 'APPLE GRAPH',figsize=(20,20),kind = 'hist') #! draws a graph for the dataframe
                          #* Legend stands for the graph key , title: gives the graph a title , figsize: changes the size of the window
```

```python
phones = phones.loc['APPLE']
phones['original_price'].plot() #! you can plot a graph for a certain column
```

```python
phones.plot(x = 'original_price',y='storage',kind = 'scatter',legend=False) 
#! you can plot with different x's and y's, also change the type to a scatterplot
```
